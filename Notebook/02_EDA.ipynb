```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Cleaned_dataset.csv')
df.head()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Aggregate rating'], bins=20)
plt.title("Distribution of Aggregate Rating")
plt.xlabel("Aggregate Rating")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='Aggregate rating')
plt.title("Count of Each Rating Value")
plt.xlabel("Aggregate Rating")
plt.ylabel("Frequency")
plt.show()
```

```python
rating_counts = df['Aggregate rating'].value_counts().sort_index()
rating_percent = (rating_counts / len(df)) * 100

print("Rating Counts:\n", rating_counts)
print("\nRating Percentages (%):\n", rating_percent)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=rating_counts.index, y=rating_percent)
plt.title("Percentage Distribution of Aggregate Ratings")
plt.xlabel("Aggregate Rating")
plt.ylabel("Percentage (%)")
plt.show()
```

```python
plt.figure(figsize=(10,5))
df['Country Code'].value_counts().plot(kind='bar')
plt.title('Top Country Codes Distribution')
plt.xlabel('Country Code')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12,6))
df['City'].value_counts().head(20).plot(kind='bar')
plt.title('Top 20 Cities with Most Restaurants')
plt.xlabel('City')
plt.ylabel('Count')
plt.show()
```

```python
cuisine_split = df['Cuisines'].str.split(', ', expand=True).stack()
cuisine_dist = cuisine_split.value_counts()
print("\n Top 20 Cuisines :\n", cuisine_dist.head(20))
```

```python
plt.figure(figsize=(12,6))
cuisine_dist.head(20).plot(kind='bar')
plt.title('Top 20 Cuisines')
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.show()
```

```python
# Spatial Visualization

plt.figure(figsize=(10,6))
plt.scatter(df['Longitude'], df['Latitude'], alpha=0.5)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Restaurant Locations on Map")
plt.show()
```

```python
# Rating vs Location

plt.figure(figsize=(8,5))
sns.scatterplot(x=df['Latitude'], y=df['Aggregate rating'])
plt.title("Latitude vs Aggregate Rating")
plt.xlabel("Latitude")
plt.ylabel("Rating")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['Longitude'], y=df['Aggregate rating'])
plt.title("Longitude vs Aggregate Rating")
plt.xlabel("Longitude")
plt.ylabel("Rating")
plt.show()
```

```python
corr_lat = df['Latitude'].corr(df['Aggregate rating'])
corr_lon = df['Longitude'].corr(df['Aggregate rating'])

print("Correlation between Latitude and Rating:", corr_lat)
print("Correlation between Longitude and Rating:", corr_lon)
```

```python
# Correlation Heatmap

num_cols = df.select_dtypes(include = [np.number]).columns
exclude_cols = ['Restaurant ID', 'Country Code', 'Latitude', 'Longitude']
num_cols = [col for col in num_cols if col not in exclude_cols]

plt.figure(figsize=(10,8))
sns.heatmap(df[num_cols].corr(), annot=False, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Outlier Detection 

outlier_rows = set()
summary = []
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    mask = (df[col] < lower) | (df[col] > upper)
    count = mask.sum()
    summary.append((col, int(count), round(count / len(df) * 100, 3)))
    outlier_rows.update(df.index[mask].tolist())

summary_df = pd.DataFrame(summary, columns=['column','outlier_count','outlier_pct']).set_index('column').sort_values('outlier_count', ascending=False)
print(summary_df)
print(f"\nRows with any numeric-column IQR outlier: {len(outlier_rows)}")
# show a few example outlier rows (if you want)
if outlier_rows:
    display(df.loc[sorted(outlier_rows)].head())
```

```python

# Outlier capping function
def cap_outliers(col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df[col] = df[col].clip(lower, upper)


for col in num_cols:
    cap_outliers(col)
```

```python
df.to_csv('Updates_dataset.csv', index = False)
print("Successfully created.")
```
